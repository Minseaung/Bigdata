```python
import requests
import openpyxl

client_id = 'c60xrhOSl1YjuvSu4PLa'
client_secret = 'HtJgQhly0x'
start, num = 1,0

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['랭킹', '제목', '링크'])

for index in range(10):
    start_number = start+(index+100)
    naver_open_api = 'https://openapi.naver.com/v1/search/news.json?query=미나리&display=100&start='+str(start_number)
    header_params = {'X-naver-client-id':client_id, 'X-naver-client-secret':client_secret}
    res = requests.get(naver_open_api, headers=header_params)

    if res.status_code == 200:
        data = res.json()
        for item in data['items']:
            num += 1
            excel_sheet.append([num, item['title'], item['link']])
    else:
        print('Error Code:', res.status_code)

excel_file.save('미나리.xlsx')
excel_file.close()
```

### XML(Extensible Markup Language)
- 특정 목적에 따라 데이터를 태그로 감싸서 마크업하는 범용적인 포멧
- 마크업 언어는 태그 등을 이용하여 데이터의 구조를 기술하는 언어의 한 가지
- 가장 친숙한 마크업 언어가 HTML
- XML은 HTML과 마찬가지로 데이터를 계층 구조로 표현
- XML 기본 구조
   - <태그 속성="속성값">내용</태그>

- 태그와 속성은 특정 목적에 따라 임의로 이름을 정해서 사용

```python
import requests
from bs4 import BeautifulSoup

service_key = 'gWYgYzsNM5isPKOi3wKYadRLfogMdwF0cJRQBZqbcssdWE2DXTgPWlGmUZdbCeeD22xElaXVRLnmeYqjcXBUpg%3D%3D'
params = '&numOfRows=10&pageNo=1&sidoName=서울'
open_api = 'http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getCtprvnRltmMesureDnsty?serviceKey='+service_key+params

res = requests.get(open_api)
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.find_all('item')
for item in data:
    #print(item)
    stationname = item.find('stationname')
    pm10grade = item.find('pm10grade')
    print(stationname.get_text(), pm10grade.get_text())
```
